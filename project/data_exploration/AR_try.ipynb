```python
from apyori import apriori
from statistic_helper import *
```

```python
students1 = load_data_from_file("studentBehaviorInfo_1.json")
students2 = load_data_from_file("studentBehaviorInfo_2.json")
students3 = load_data_from_file("studentBehaviorInfo_all.json")
students = students1 + students2 + students3

crawled_video_info = load_data_from_file('data_crawled_with_youtube_info_array.json')
```

```python
# load attr 'chosenVideo'
sList = []
for student in students3:
    
    if 3913 in student['chosenVideo']:
        student['chosenVideo'].remove(3913)
        
    sList.append(student['chosenVideo'])

print sList
```

```python
# run Apriori 
# params could be min_support, min_confidence, max_length
results = list(apriori(sList, min_support=0.21))
for result in results:
    print result
```

```python
for result in results:
    print result[0]
```

```python
video_info_dict = {v['postId']: v for v in crawled_video_info}
```

```python
# print video titles of rules
for result in results:
    # catch data in frozenset
    for frozen_item in result[0]:
        if unicode(frozen_item) in video_info_dict:
            video = video_info_dict[unicode(frozen_item)]
            print video['title']
```
