```python
crawl_listing = open('crawl_listing', 'r')
urls = crawl_listing.readlines()
data = {}
for url in urls:
    print url.strip()
    url_extract = Extractor(url)
    data[url_extract.post_id] = {}
    data[url_extract.post_id]['youtube_link'] = url_extract.youtube_link
    data[url_extract.post_id]['wordList'] = url_extract.word_list
    
```
